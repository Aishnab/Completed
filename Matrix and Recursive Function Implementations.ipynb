Question 1<br>
(5 marks)<br>
Create a python script that will prompt the user for 9 values, checking that they are all valid numbers,
and populate those given values into a 3x3 Matrix called A.
The program should present the created matrix , the determinant of the matrix and, if possible, the
inverse of the matrix.
The program should also present the results of the matrix operations A*A and A*A*A.<br>
Finally the program should print the number of values in the matrix that >= 10
You will need to evidence the outputs, as well as the error handling of any invalid values when the
user attempts to input them in the matrix.

```python
import numpy as np

def prompt_user_for_matrix_values(matrix_size):
    """
    Prompts the user for values to populate a matrix of a given size.
    
    Args:
    - matrix_size: a tuple representing the size of the matrix as (num_rows, num_columns)
    
    Returns:
    - A numpy array representing the matrix created from user input
    """
    matrix = np.zeros(matrix_size)
    for i in range(matrix_size[0]):
        for j in range(matrix_size[1]):
            while True:
                try:
                    value = float(input(f"Enter a value for row {i+1}, column {j+1}: "))
                    break
                except ValueError:
                    print("Invalid input. Please enter a valid number.")
            matrix[i][j] = value
    return matrix

def print_matrix_info(matrix):
    """
    Prints information about a matrix including the matrix itself, its determinant, its inverse (if possible), the result of
    A*A and A*A*A, and the number of values in the matrix that are >= 10.
    
    Args:
    - matrix: a numpy array representing the matrix to be analyzed
    """
    print(f"Matrix:\n{matrix}")
    determinant = np.linalg.det(matrix)
    print(f"Determinant: {determinant}")
    if determinant != 0:
        inverse = np.linalg.inv(matrix)
        print(f"Inverse:\n{inverse}")
    else:
        print("Matrix is not invertible.")
    A_squared = np.matmul(matrix, matrix)
    A_cubed = np.matmul(A_squared, matrix)
    print(f"A*A:\n{A_squared}\nA*A*A:\n{A_cubed}")
    count = np.sum(matrix >= 10)
    print(f"Number of values in matrix that are >= 10: {count}")

# prompt the user for a 3x3 matrix
A = prompt_user_for_matrix_values((3,3))

# analyze the matrix
print_matrix_info(A)

```

Question 3<br>
(5 marks)<br>
Ackermann's function (A) is defined as:
A(m,n):<br>
n+1, if m=0<br>
A(m-1,1), if m>0 and n=0<br>
A(m-1,A(m,n-1)), if m>0 and n> 0<br>
Write a recursive Python function to implement Ackermann's function. How many recursive calls will
be required to evaluate A(2,3)?



```python
def Ackermann(m,n):
  if m == 0:
    return n+1
  elif n==0:
    return Ackermann(m-1,10)
  else:
    return Ackermann(m-1,Ackermann(m,n-1))
print(Ackermann(2,3))
```
